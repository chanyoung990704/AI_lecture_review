행렬 계산을 통해 간단한 신경망 Skeleton 구현

```python
input = np.asarray([0.9, 0.1, 0.8]) # 인풋 배열을 받아 행렬로 바꿈
weight_input_layer1 = np.asarray(
    [
        [0.9, 0.2, 0.1],
        [0.3, 0.8, 0.5],
        [0.4, 0.2, 0.6]
    ]
)
print(weight_input_layer1)
print("============")

weight_layer1_layer2 = np.asarray([
    [0.3, 0.6, 0.8],
    [0.7, 0.5, 0.1],
    [0.5, 0.2, 0.9]

])

print(weight_layer1_layer2)
```

```python
input_layer1 = np.dot(input, weight_input_layer1)
print(input_layer1)
```

시그모이드 함수를 통해 값 변환

```python
import scipy.special
sigmoid_function = lambda x : scipy.special.expit(x)

input_layer1_sig = sigmoid_function(input_layer1)
print(input_layer1_sig)
```

변환한 값을 인풋 변수로 사용

```python
layer1_layer2 = np.dot(input_layer1_sig, weight_layer1_layer2)
print(layer1_layer2)

print("=======")
layer1_layer2_sig = sigmoid_function(layer1_layer2)
print(layer1_layer2_sig)
```
